```python
df[(df["yr"]==0)&(df["mnth"]==1)]["temp"].mean()
```

```python
# multiple group columns and multiple aggrgate columns
df.groupby(by=["yr","mnth"])[["temp","windspeed"]].mean()
```

```python
# multiple aggregate functions
df.groupby(by="mnth").agg({"temp":"mean","windspeed":"median"})
```

```python
#df[df["mnth"]==1]["temp"].mean()
df[df["mnth"]==1]["windspeed"].median()
```

```python
# multilple group columns, multiple aggregate column and multiple aggregate functions
df.groupby(by=["yr","mnth"]).agg({"temp":"mean","windspeed":"median"})
```

```python
# Complicated Combination
# Multiple group columns, Multiple aggregate columns, Multiple aggregate functions 
#on same columns

df.groupby(by=["yr","mnth"]).agg({"temp":["min","max","mean"],"windspeed":["std","var"]})
```

## Concat

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
```

```python
frames = [df1, df2, df3]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
result1 = pd.concat(frames,axis=1)
result1
```

## Merging

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
left
```

```python
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
right
```

```python
result = pd.merge(left, right, on="key",how='left')
result
```

## Pivot Table

```python
df = pd.DataFrame({'ID': {0: 23, 1: 43, 2: 12, 
                          3: 13, 4: 67, 5: 89,
                          6: 90, 7: 56, 8: 34},
                     
                 'Name': {0: 'Ram', 1: 'Deep', 2: 'Yash',
                          3: 'Aman', 4: 'Arjun', 5: 'Aditya',
                          6: 'Akash', 7: 'Chalsea',
                          8: 'Divya'},
                     
                 'Marks': {0: 89, 1: 97, 2: 45,
                           3: 78, 4: 56, 5: 76,
                           6: 81, 7: 87, 8: 100},
                     
                 'Grade': {0: 'B', 1: 'A', 2: 'F',
                           3: 'C', 4: 'E', 5: 'C',
                           6: 'B', 7: 'B', 8: 'A'}})
  
# simple use pivot_table() method
print(pd.pivot_table(df, index = ["ID"]))
```
